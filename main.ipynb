Importing libraries and setting up connection

````python
from pymongo import MongoClient
import pandas as pd
import time
client = MongoClient("localhost:27017")
````

````python
db = client["Ironhack"]
c = db.get_collection("companies")
````

### 1. All the companies whose name match 'Babelgum'. Retrieve only their `name` field.

````python
result =list(c.find({'name': 'Babelgum'}, {'name': 1, '_id': 0}))
result
````

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

````python
filters = {'number_of_employees': {'$gt': 5000}}
projection = {'name': 1}

result2 = list(c.find(filters, projection).sort("number_of_employees", -1).limit(20))
result2
````

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fields.

````python
filters = {'founded_year': {'$gte': 2000, '$lte': 2005}}
projection = {'name': 1, '_id': 0, 'founded_year': 1}

result3 = list(c.find(filters, projection).sort("founded_year", -1))
result3
````

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

````python
filters = {'ipo.valuation_amount': {'$gt': 100000000}, 'founded_year': {'$lt': 2010}}
projection = {'name': 1, '_id': 0, 'ipo': 1}

result4 = list(c.find(filters, projection).sort("ipo.valuation_amount", -1))
result4
````

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

````python
filters = {'number_of_employees': {'$lt': 1000}, 'founded_year': {'$lt': 2005}}
projection = {'name': 1}

result5 = list(c.find(filters, projection).sort("number_of_employees", -1).limit(10))
result5
````

### 6. All the companies that don't include the `partners` field.

````python
filters = {'partners': {'$exists': 'False'}}
projection = {'name': 1, '_id': 0, 'partners': 1}

result6 = list(c.find(filters, projection))
result6
````

### 7. All the companies that have a null type of value on the `category_code` field.

````python
filters = {'category_code': {'$type': 10}}
projection = {'name': 1, '_id': 0, 'category_code': 1}

result7 = list(c.find(filters, projection))
result7
````

### 8. All the companies that have at least 100 employees but less than 1000. Retrieve only the `name` and `number of employees` fields.

````python
filters = {'number_of_employees': {'$gte': 100, '$lt': 1000}}
projection = {'name': 1, 'number_of_employees': 1, '_id': 0}

result8 = list(c.find(filters, projection).sort("number_of_employees", -1))
result8
````

### 9. Order all the companies by their IPO price in a descending order.

````python
projection = {'name': 1, '_id': 0, 'ipo.valuation_amount': 1}

result9 = list(c.find({}, projection).sort("ipo.valuation_amount", -1))
result9
````

### 10. Retrieve the 10 companies with more employees, order by the `number of employees`

````python
projection = {'name': 1, 'number_of_employees': 1, '_id': 0}

result10 = list(c.find({}, projection).sort("number_of_employees", -1).limit(10))
result10
````

### 11. All the companies founded on the second semester of the year. Limit your search to 1000 companies.

````python
filters = {'founded_month': {'$gt': 6}}
projection = {'name': 1, 'founded_month': 1, '_id': 0, 'founded_year': 1}

result11 = list(c.find(filters, projection).sort("founded_year", 1).limit(1000))
result11
````

### 12. All the companies founded before 2000 that have an acquisition amount of more than 10.000.000

````python
filters = {'founded_year': {'$lt': 2000}, 'acquisition.price_amount': {'$gt': 10000000}}
projection = {'name': 1, '_id': 0, 'founded_year': 1, 'acquisition.price_amount': 1}

result12 = list(c.find(filters, projection).sort("acquisition.price_amount", -1))
result12
````

### 13. All the companies that have been acquired after 2010, order by the acquisition amount, and retrieve only their `name` and `acquisition` field.

````python
filters = {'acquisition.acquired_year': {'$gt': 2010}}
projection = {'name': 1, '_id': 0, 'acquisition': 1}

result13 = list(c.find(filters, projection).sort("acquisition.price_amount", -1))
result13
````

### 14. Order the companies by their `founded year`, retrieving only their `name` and `founded year`.

````python
projection = {'name': 1, '_id': 0, 'founded_year': 1}

result14 = list(c.find({}, projection).sort("founded_year", -1))
result14
````

### 15. All the companies that have been founded on the first seven days of the month, including the seventh. Sort them by their `acquisition price` in a descending order. Limit the search to 10 documents.

````python
filters = {'founded_day': {'$lte': 7}}
projection = {'name': 1, '_id': 0, 'founded_day': 1}

result15 = list(c.find(filters, projection).sort("acquisition.price_amount", -1).limit(10))
result15
````

### 16. All the companies on the 'web' `category` that have more than 4000 employees. Sort them by the amount of employees in ascending order.

````python
filters = {'category_code': 'web', 'number_of_employees': {'$gt': 4000}}
projection = {'name': 1, '_id': 0, 'number_of_employees': 1, 'category_code': 1}

result16 = list(c.find(filters, projection).sort("number_of_employees", 1))
result16
````

### 17. All the companies whose acquisition amount is more than 10.000.000, and currency is 'EUR'.

````python
filters = {'acquisition.price_currency_code': 'EUR', 'acquisition.price_amount': {'$gt': 10000000}}
projection = {'name': 1, '_id': 0, 'acquisition.price_amount': 1}

result17 = list(c.find(filters, projection).sort("acquisition.price_amount", -1))
result17
````

### 18. All the companies that have been acquired on the first trimester of the year. Limit the search to 10 companies, and retrieve only their `name` and `acquisition` fields.

````python
filters = {'acquisition.acquired_month': {'$lte': 3}}
projection = {'name': 1, '_id': 0, 'acquisition': 1}

result18 = list(c.find(filters, projection).sort("acquisition.acquired_year", -1).limit(10))
result18
````
